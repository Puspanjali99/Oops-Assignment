# 1: Bank Account Create a class representing a bank account with attributes like account number, account holder name, and balance. Implement methods to deposit and withdraw money from the account.

```python
class BankAccount:
    def __init__(self, account_number, account_holder_name, balance=0.0):
        self.account_number = account_number
        self.account_holder_name = account_holder_name
        self.balance = balance

    def deposit(self, amount):
        if amount > 0:
            self.balance += amount
            print(f"Deposited ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid deposit amount. Please deposit a positive amount.")

    def withdraw(self, amount):
        if amount > 0 and amount <= self.balance:
            self.balance -= amount
            print(f"Withdrew ${amount}. New balance: ${self.balance}")
        else:
            print("Invalid withdrawal amount or insufficient funds.")

    def get_balance(self):
        return self.balance

    def display_account_info(self):
        print(f"Account Number: {self.account_number}")
        print(f"Account Holder: {self.account_holder_name}")
        print(f"Balance: ${self.balance}")



account1 = BankAccount(account_number="123456789", account_holder_name="John Doe", balance=1000.0)
account1.display_account_info()

account1.deposit(500.0)
account1.withdraw(200.0)
account1.display_account_info()
```

# 2: Employee Management Create a class representing an employee with attributes like employee ID, name, and salary. Implement methods to calculate the yearly bonus and display employee details.

```python
class Employee:
    def __init__(self, employee_id, name, salary):
        self.employee_id = employee_id
        self.name = name
        self.salary = salary

    def calculate_yearly_bonus(self, bonus_percentage):
        if 0 <= bonus_percentage <= 100:
            bonus_amount = (bonus_percentage / 100) * self.salary
            return bonus_amount
        else:
            print("Invalid bonus percentage. Please provide a percentage between 0 and 100.")
            return 0

    def display_employee_details(self):
        print(f"Employee ID: {self.employee_id}")
        print(f"Name: {self.name}")
        print(f"Salary: ${self.salary:.2f}")


employee1 = Employee(employee_id=1, name="Alice", salary=60000.0)
employee1.display_employee_details()

bonus_percentage = 10 
bonus_amount = employee1.calculate_yearly_bonus(bonus_percentage)
print(f"Yearly Bonus ({bonus_percentage}%): ${bonus_amount:.2f}")


employee1.salary += bonus_amount
employee1.display_employee_details()
```

# 3: Vehicle Rental Create a class representing a vehicle rental system. Implement methods to rent a vehicle, return a vehicle, and display available vehicles.

```python
class VehicleRentalSystem:
    def __init__(self):
        self.available_vehicles = {}  
    def add_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            print(f"Added {vehicle_type} with ID {vehicle_id} to available vehicles.")
        else:
            print(f"Vehicle with ID {vehicle_id} already exists in the system.")

    def rent_vehicle(self, vehicle_id):
        if vehicle_id in self.available_vehicles:
            rented_vehicle_type = self.available_vehicles.pop(vehicle_id)
            print(f"Rented {rented_vehicle_type} with ID {vehicle_id}.")
        else:
            print(f"Vehicle with ID {vehicle_id} is not available for rent.")

    def return_vehicle(self, vehicle_id, vehicle_type):
        if vehicle_id not in self.available_vehicles:
            self.available_vehicles[vehicle_id] = vehicle_type
            print(f"Returned {vehicle_type} with ID {vehicle_id}.")
        else:
            print(f"Vehicle with ID {vehicle_id} is already available.")

    def display_available_vehicles(self):
        if self.available_vehicles:
            print("Available Vehicles:")
            for vehicle_id, vehicle_type in self.available_vehicles.items():
                print(f"ID: {vehicle_id}, Type: {vehicle_type}")
        else:
            print("No vehicles available.")

rental_system = VehicleRentalSystem()

rental_system.add_vehicle(vehicle_id=1, vehicle_type="Sedan")
rental_system.add_vehicle(vehicle_id=2, vehicle_type="SUV")
rental_system.add_vehicle(vehicle_id=3, vehicle_type="Motorcycle")

rental_system.display_available_vehicles()

rental_system.rent_vehicle(vehicle_id=2)

rental_system.display_available_vehicles()

rental_system.return_vehicle(vehicle_id=2, vehicle_type="SUV")

rental_system.display_available_vehicles()
```

# 4: Library Catalog Create classes representing a library and a book. Implement methods to add books to the library, borrow books, and display available books.

```python
class Book:
    def __init__(self, book_id, title, author):
        self.book_id = book_id
        self.title = title
        self.author = author
        self.is_borrowed = False

    def borrow(self):
        if not self.is_borrowed:
            self.is_borrowed = True
            print(f"Borrowed '{self.title}' by {self.author}.")
        else:
            print(f"'{self.title}' is already borrowed.")

    def return_book(self):
        if self.is_borrowed:
            self.is_borrowed = False
            print(f"Returned '{self.title}' by {self.author}.")
        else:
            print(f"'{self.title}' is not currently borrowed.")

    def display_info(self):
        print(f"Book ID: {self.book_id}")
        print(f"Title: {self.title}")
        print(f"Author: {self.author}")
        print("Status: Borrowed" if self.is_borrowed else "Status: Available")
        print()


class Library:
    def __init__(self):
        self.books = {}

    def add_book(self, book):
        if book.book_id not in self.books:
            self.books[book.book_id] = book
            print(f"Added '{book.title}' by {book.author} to the library.")
        else:
            print(f"Book with ID {book.book_id} already exists in the library.")

    def borrow_book(self, book_id):
        if book_id in self.books:
            self.books[book_id].borrow()
        else:
            print(f"Book with ID {book_id} is not available in the library.")

    def return_book(self, book_id):
        if book_id in self.books:
            self.books[book_id].return_book()
        else:
            print(f"Book with ID {book_id} is not in the library.")

    def display_available_books(self):
        available_books = [book for book in self.books.values() if not book.is_borrowed]
        if available_books:
            print("Available Books:")
            for book in available_books:
                book.display_info()
        else:
            print("No available books in the library.")



library = Library()

book1 = Book(book_id=1, title="The Great Gatsby", author="F. Scott Fitzgerald")
book2 = Book(book_id=2, title="To Kill a Mockingbird", author="Harper Lee")
book3 = Book(book_id=3, title="1984", author="George Orwell")

library.add_book(book1)
library.add_book(book2)
library.add_book(book3)

library.display_available_books()

library.borrow_book(2)
library.borrow_book(3)
library.borrow_book(1)

library.display_available_books()

library.return_book(2)

library.display_available_books()
```

#  5: Product Inventory Create classes representing a product and an inventory system. Implement methods to add products to the inventory, update product quantity, and display available products.

```python
class Product:
    def __init__(self, product_id, name, price, quantity):
        self.product_id = product_id
        self.name = name
        self.price = price
        self.quantity = quantity

    def update_quantity(self, new_quantity):
        if new_quantity >= 0:
            self.quantity = new_quantity
            print(f"Updated quantity of {self.name} to {new_quantity}.")
        else:
            print("Invalid quantity. Please provide a non-negative value.")

    def display_info(self):
        print(f"Product ID: {self.product_id}")
        print(f"Name: {self.name}")
        print(f"Price: ${self.price:.2f}")
        print(f"Quantity: {self.quantity}")
        print()


class InventorySystem:
    def __init__(self):
        self.products = {}

    def add_product(self, product):
        if product.product_id not in self.products:
            self.products[product.product_id] = product
            print(f"Added '{product.name}' to the inventory.")
        else:
            print(f"Product with ID {product.product_id} already exists in the inventory.")

    def update_quantity(self, product_id, new_quantity):
        if product_id in self.products:
            self.products[product_id].update_quantity(new_quantity)
        else:
            print(f"Product with ID {product_id} is not in the inventory.")

    def display_available_products(self):
        if self.products:
            print("Available Products:")
            for product in self.products.values():
                product.display_info()
        else:
            print("No products in the inventory.")


inventory_system = InventorySystem()

product1 = Product(product_id=1, name="Laptop", price=999.99, quantity=10)
product2 = Product(product_id=2, name="Smartphone", price=499.99, quantity=20)
product3 = Product(product_id=3, name="Headphones", price=49.99, quantity=30)

inventory_system.add_product(product1)
inventory_system.add_product(product2)
inventory_system.add_product(product3)

inventory_system.display_available_products()

inventory_system.update_quantity(product_id=2, new_quantity=15)

inventory_system.display_available_products()
```

#  6: Shape Calculation Create a class representing a shape with attributes like length, width, and height. Implement methods to calculate the area and perimeter of the shape.

```python
class Shape:
    def __init__(self, length, width=None, height=None):
        self.length = length
        self.width = width
        self.height = height

    def calculate_area(self):
        if self.width is not None and self.height is not None:
            # For rectangle or square
            return self.length * self.width
        elif self.width is not None:
            # For square or circle
            return self.length * self.length
        else:
            print("Cannot calculate area. Insufficient information.")

    def calculate_perimeter(self):
        if self.width is not None and self.height is not None:
            # For rectangle
            return 2 * (self.length + self.width)
        elif self.width is not None:
            # For square or circle
            return 4 * self.length
        else:
            print("Cannot calculate perimeter. Insufficient information.")



rectangle = Shape(length=5, width=3)
print(f"Rectangle Area: {rectangle.calculate_area()}")
print(f"Rectangle Perimeter: {rectangle.calculate_perimeter()}")

square = Shape(length=4)
print(f"Square Area: {square.calculate_area()}")
print(f"Square Perimeter: {square.calculate_perimeter()}")
```

# 7: Student Management Create a class representing a student with attributes like student ID, name, and grades. Implement methods to calculate the average grade and display student details.

```python
class Student:
    def __init__(self, student_id, name, grades):
        self.student_id = student_id
        self.name = name
        self.grades = grades

    def calculate_average_grade(self):
        if len(self.grades) > 0:
            average_grade = sum(self.grades) / len(self.grades)
            return average_grade
        else:
            print("Cannot calculate average grade. No grades available.")

    def display_student_details(self):
        print(f"Student ID: {self.student_id}")
        print(f"Name: {self.name}")
        if self.grades:
            print("Grades:", ", ".join(map(str, self.grades)))
            print(f"Average Grade: {self.calculate_average_grade()}")
        else:
            print("No grades available.")



student1 = Student(student_id="S001", name="Alice", grades=[85, 90, 78, 92, 88])
student1.display_student_details()

student2 = Student(student_id="S002", name="Bob", grades=[75, 88, 92, 79, 85])
student2.display_student_details()

student3 = Student(student_id="S003", name="Charlie", grades=[])
student3.display_student_details()
```

# 8: Email Management Create a class representing an email with attributes like sender, recipient, and subject. Implement methods to send an email and display email details.

```python
class Email:
    def __init__(self, sender, recipient, subject, body):
        self.sender = sender
        self.recipient = recipient
        self.subject = subject
        self.body = body
        self.is_sent = False

    def send_email(self):
        if not self.is_sent:
           
            print(f"Email sent from {self.sender} to {self.recipient}.")
            self.is_sent = True
        else:
            print("Email has already been sent.")

    def display_email_details(self):
        print(f"Sender: {self.sender}")
        print(f"Recipient: {self.recipient}")
        print(f"Subject: {self.subject}")
        print("Body:")
        print(self.body)
        print(f"Status: {'Sent' if self.is_sent else 'Not Sent'}")



email1 = Email(sender="alice@example.com", recipient="bob@example.com", subject="Meeting Details", body="Hello Bob, let's meet tomorrow at 2 PM.")
email1.display_email_details()

email1.send_email()
email1.display_email_details()

email2 = Email(sender="charlie@example.com", recipient="dave@example.com", subject="Project Update", body="Hi Dave, here is the latest project update.")
email2.display_email_details()

email2.send_email()
email2.display_email_details()
```

# 9: Social Media Profile Create a class representing a social media profile with attributes like username and posts. Implement methods to add posts, display posts, and search for posts by keyword.

```python
class SocialMediaProfile:
    def __init__(self, username):
        self.username = username
        self.posts = []

    def add_post(self, post_content):
        new_post = {"content": post_content, "author": self.username}
        self.posts.append(new_post)
        print(f"New post added by {self.username}.")

    def display_posts(self):
        if self.posts:
            print(f"Posts by {self.username}:")
            for index, post in enumerate(self.posts, start=1):
                print(f"{index}. {post['content']} - {post['author']}")
        else:
            print(f"No posts by {self.username}.")

    def search_posts(self, keyword):
        matching_posts = [post for post in self.posts if keyword.lower() in post['content'].lower()]
        if matching_posts:
            print(f"Matching posts for '{keyword}' by {self.username}:")
            for index, post in enumerate(matching_posts, start=1):
                print(f"{index}. {post['content']} - {post['author']}")
        else:
            print(f"No matching posts for '{keyword}' by {self.username}.")



profile1 = SocialMediaProfile(username="Alice123")
profile1.add_post("Enjoying a sunny day at the beach!")
profile1.add_post("Just finished reading a great book.")
profile1.display_posts()

profile1.search_posts("book")

profile2 = SocialMediaProfile(username="Bob456")
profile2.add_post("Excited for the upcoming project deadline.")
profile2.add_post("Trying out a new recipe for dinner.")
profile2.display_posts()

profile2.search_posts("deadline")
```

# 10: ToDo List Create a class representing a ToDo list with attributes like tasks and due dates. Implement methods to add tasks, mark tasks as completed, and display pending tasks.

```python
class ToDoList:
    def __init__(self):
        self.tasks = []

    def add_task(self, task, due_date=None):
        new_task = {"task": task, "due_date": due_date, "completed": False}
        self.tasks.append(new_task)
        print(f"New task added: {task}")

    def mark_task_as_completed(self, task_index):
        if 0 <= task_index < len(self.tasks):
            self.tasks[task_index]["completed"] = True
            print(f"Task '{self.tasks[task_index]['task']}' marked as completed.")
        else:
            print("Invalid task index. Please provide a valid index.")

    def display_pending_tasks(self):
        pending_tasks = [task for task in self.tasks if not task["completed"]]
        if pending_tasks:
            print("Pending Tasks:")
            for index, task in enumerate(pending_tasks, start=1):
                print(f"{index}. {task['task']} (Due Date: {task['due_date']})")
        else:
            print("No pending tasks.")


todo_list = ToDoList()

todo_list.add_task("Finish report for work", due_date="2023-12-01")
todo_list.add_task("Buy groceries")
todo_list.add_task("Exercise", due_date="2023-11-15")

todo_list.display_pending_tasks()

todo_list.mark_task_as_completed(2)

todo_list.display_pending_tasks()
```
